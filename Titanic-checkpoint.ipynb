```python
survival_by_gender = df.groupby('Sex')['Survived'].sum()
total_by_gender = df['Sex'].value_counts()
survival_rate = (survival_by_gender / total_by_gender) * 100

print("\nSurvival counts by gender:")
print(survival_by_gender)
print("\nSurvival rate by gender (%):")
print(survival_rate)
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].sum()
total_by_class = df.groupby('Pclass')['Survived'].count()
survival_rate_by_class = (survival_by_class / total_by_class) * 100

print("Survival by Passenger Class:")
print(pd.DataFrame({
    'Survived': survival_by_class,
    'Total': total_by_class,
    'Survival Rate (%)': survival_rate_by_class
}))
```
